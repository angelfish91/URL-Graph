```python
def get_keywords():
    with open("./src/keyword.txt", "r") as f:
        keywords = f.read()
        keywords = [_.strip() for _ in keywords.strip().split(',')]
    return keywords 
```

```python
keywords = get_keywords()
```

```python
df = pd.read_csv("../data/sangfor/safe_0419.csv")
urls = list(df.url)
```

```python
detect = set()
for url in urls:
    host = urlparse.urlparse(url).hostname
    tld = tldextract.extract(host)
    for key in keywords:
        if key in tld.domain:
            detect.add(host)
```
